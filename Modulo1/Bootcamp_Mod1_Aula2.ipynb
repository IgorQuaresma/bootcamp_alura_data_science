**Analisando Dados da Saúde com Python e Pandas**

```python
import pandas as pd
```

```python
dados = pd.read_csv("/A222426189_28_143_208.csv", encoding="ISO-8859-1",
                    skiprows = 3, sep=";", skipfooter=12,
                    thousands=".", decimal=",")

dados.head()
```

**Aula 2**

```python
colunas_preenchidas = dados.mean().index.tolist()
colunas_preenchidas.insert(0, "Unidade da Federação")
colunas_preenchidas
```

```python
usaveis = dados[colunas_preenchidas]
```

```python
usaveis = usaveis.set_index("Unidade da Federação")
```

```python
usaveis.T.plot()
```

```python
usaveis = usaveis.drop("Total", axis=1)
usaveis
```

```python
usaveis.T.plot(figsize=(10,6))
```

Desafio 01: Reposicionar a legenda do gráfico em uma posição mais adequada



```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
#Solução utilizando a legenda dentro do frame do gráfico

ax = usaveis.T.plot(figsize=(10,6), title="Gastos com atendimentos pelo SUS por Estado")
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.2f}"))
ax.legend(loc="upper center", fontsize="small", ncol=4, bbox_to_anchor=(0.55,1))
plt.show()
```

```python
#Solução utilizando a movimentação da legenda para fora do frame do gráfico 

ax = usaveis.T.plot(figsize=(10,6), title="Gastos com atendimentos pelo SUS por Estado")
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.2f}"))
ax.legend(loc="upper right", fontsize="small", bbox_to_anchor=(0.78, 0.58,0.5, 0.5))
plt.show()
```


Desafio 02: Plotar o gráfico de linha com apenas 5 estados de sua preferência

```python
#Solução utilizando a legenda dentro do frame do gráfico

colunas_selecionadas = ['53 Distrito Federal','13 Amazonas','23 Ceará','31 Minas Gerais','43 Rio Grande do Sul']
usaveis2 = usaveis.loc[colunas_selecionadas,:].T

ax = usaveis2.plot(figsize=(10,6), title="Gastos com atendimentos pelo SUS por Estado")
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.2f}"))
ax.legend(loc="upper center", fontsize="small", ncol=4)
plt.show()
```

```python
#Solução utilizando a movimentação da legenda para fora do frame do gráfico 

colunas_selecionadas = ['53 Distrito Federal','13 Amazonas','23 Ceará','31 Minas Gerais','43 Rio Grande do Sul']
usaveis2 = usaveis.loc[colunas_selecionadas,:].T

ax = usaveis2.plot(figsize=(10,6), title="Gastos com atendimentos pelo SUS por Estado")
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.2f}"))
ax.legend(loc="upper center", fontsize="small", bbox_to_anchor=((0.88, -0.25,0.5, 0.5)))
plt.show()
```
