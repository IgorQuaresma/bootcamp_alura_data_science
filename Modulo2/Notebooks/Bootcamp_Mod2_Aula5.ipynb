````python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd
import seaborn as sns

ordenado_por_total = pd.read_csv('https://raw.githubusercontent.com/IgorQuaresma/bootcamp_alura_data_science/main/Modulo2/Dados/gastos_ordenado_por_total.csv')

ordenado_por_total = ordenado_por_total.rename(columns = {'Unidade da Federação':'UF'})
ordenado_por_total = ordenado_por_total.set_index('UF')
ordenado_por_total.index = [valor[2:].strip() for valor in ordenado_por_total.index]
ordenado_por_total

ordenado_por_total.head()

mensal = ordenado_por_total.T
mensal.head()
````

````python
from datetime import date

meses = {
    "Jan" : 1,
    "Fev" : 2,
    "Mar" : 3,
    "Abr" : 4,
    "Mai" : 5,
    "Jun" : 6,
    "Jul" : 7,
    "Ago" : 8,
    "Set" : 9,
    "Out" : 10,
    "Nov" : 11,
    "Dez" : 12
}

def para_dia(ano_mes: str):
    ano: int = int(ano_mes[:4])
    mes: str = ano_mes[5:]
    mes_numero: int = meses[mes]
    return date(ano, mes_numero, 1)

mensal.index = mensal.index.map(para_dia)
mensal.head()
````

# Desafio 02: Pegar estados de climas diferentes (por exemplo, norte e sul) e plotar o gráfico de linhas, comparar ambos e compartilhar suas análises.

````python


#Criando o grafico
fig, ax = plt.subplots(1, 2, figsize=(20,8), sharex=False, sharey=True) 


#Criando o gráfico
sns.lineplot(data=mensal_aberto.query("uf=='Ceará'"), x="mes", y="gasto_diario", hue="ano", markers=True, ax=ax[0])
sns.lineplot(data=mensal_aberto.query("uf=='Rio Grande do Sul'"), x="mes", y="gasto_diario", hue="ano", ax=ax[1])
fig.suptitle("Gastos hospitalares nos estados do Ceará e Rio Grande do Sul", fontsize=16, fontweight="bold")

#Formatando
ax[0].set_title('Ceará', fontsize=14)
ax[1].set_title('Rio Grande do Sul', fontsize=14)
ax[0].get_legend().remove()
ax[1].legend(bbox_to_anchor=(-0.52, 0.5, 0.5, 0.5), fontsize='large')
ax[0].set_ylabel('Gastos', fontsize=12)
ax[0].set_xlabel('Mês', fontsize=12)
ax[1].set_xlabel('Mês', fontsize=12)
ax[0].grid(linestyle="--")
ax[1].grid(linestyle="--")
ax[0].yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.0f} Milhões"))

plt.ylabel('Gastos')
plt.show()
````
