# Desafio: Analise os gráficos e tente chegar a hipóteses que explicam o que você está entendendo dos dados. Lembre-se que a conclusão é uma hipótese. E que esses dados serão alterados.

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd
import seaborn as sns
```

```python
ordenado_por_total = pd.read_csv('https://raw.githubusercontent.com/IgorQuaresma/bootcamp_alura_data_science/main/Modulo2/Dados/gastos_ordenado_por_total.csv')
```

```python
ordenado_por_total = ordenado_por_total.rename(columns = {'Unidade da Federação':'UF'})
ordenado_por_total = ordenado_por_total.set_index('UF')
ordenado_por_total.index = [valor[2:].strip() for valor in ordenado_por_total.index]
ordenado_por_total
```

```python
ordenado_por_total.head()
```

```python
mensal = ordenado_por_total.T
mensal.head()
```

# Desafio 01: Fazer o scatter plot feito em aula para o seu estado, leia um pouco da documentação sobre este gráfico e teste alguns parâmetros.

```python
sns.set_style('darkgrid')

fig, ax = plt.subplots(figsize=(10,8))

sns.scatterplot(data=mensal, x=mensal.index, y="Distrito Federal")
plt.title('Série histórica de gastos de saúde do SUS no DF', fontweight='bold', fontsize=15)
plt.xlabel('Ano/Mês', fontsize=12)
plt.ylabel('Valor Total', fontsize=12)
plt.xticks(rotation=45)
ax.xaxis.set_major_locator(ticker.IndexLocator(base=12, offset=-1))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.0f} Milhões"))
plt.show()
```

# Desafio 02: Formatar o eixo y colocando a escala em milhões usando o StrMethodFormatter, colocar um título e colocar o título nos 2 eixos, alterando os tamanhos dos títulos.


```python
sns.set_style('darkgrid')

fig, ax = plt.subplots(figsize=(10,8))

sns.lineplot(data=mensal, x=mensal.index, y="Distrito Federal")
plt.title('Série histórica de gastos de saúde do SUS no DF', fontweight='bold', fontsize=15)
plt.xlabel('Ano/Mês', fontsize=12)
plt.ylabel('Valor Total', fontsize=12)
plt.xticks(rotation=45)
plt.ylim(0, 30)
plt.grid(linestyle="--")
ax.xaxis.set_major_locator(ticker.IndexLocator(base=12, offset=-1))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.0f} Milhões"))
plt.show()
```
