````python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd
import seaborn as sns

ordenado_por_total = pd.read_csv('https://raw.githubusercontent.com/IgorQuaresma/bootcamp_alura_data_science/main/Modulo2/Dados/gastos_ordenado_por_total.csv')

ordenado_por_total = ordenado_por_total.rename(columns = {'Unidade da Federação':'UF'})
ordenado_por_total = ordenado_por_total.set_index('UF')
ordenado_por_total.index = [valor[2:].strip() for valor in ordenado_por_total.index]
ordenado_por_total

ordenado_por_total.head()

mensal = ordenado_por_total.T
mensal.head()
````

````python
from datetime import date

meses = {
    "Jan" : 1,
    "Fev" : 2,
    "Mar" : 3,
    "Abr" : 4,
    "Mai" : 5,
    "Jun" : 6,
    "Jul" : 7,
    "Ago" : 8,
    "Set" : 9,
    "Out" : 10,
    "Nov" : 11,
    "Dez" : 12
}

def para_dia(ano_mes: str):
    ano: int = int(ano_mes[:4])
    mes: str = ano_mes[5:]
    mes_numero: int = meses[mes]
    return date(ano, mes_numero, 1)

mensal.index = mensal.index.map(para_dia)
mensal.head()
````

````python
plt.figure(figsize=(10,6))
ax = sns.lineplot(data=mensal, x=mensal.index, y="São Paulo")
plt.xticks(rotation=30)
plt.ylim(0, 350)
plt.grid(linestyle="--")
plt.show()
````

````python
from datetime import datetime, timedelta, timezone

#Calcular  tempo de vida de uma determinada pessoa

#04/10/1996

def tempo_vida(data_nascimento):
    ano = int(data_nascimento.split('/')[-1])
    mes = int(data_nascimento.split('/')[1])
    dia = int(data_nascimento.split('/')[0])
    data_nascimento = datetime(ano, mes, dia)
    hoje = datetime.now() #mostra o tempo atual
    resultado = hoje - data_nascimento
    return f'Você tem {resultado}' 
````

````python
def dias_formatados():
    hoje = datetime.now()
    data = hoje.strftime('Dia da semana: %a, mes %m, dia %d de %Y')
    return data
````

````python
#escolher estados da Regiao Centro Oeste
estados_co = ['Distrito Federal', 'Mato Grosso', 'Mato Grosso do Sul', 'Goiás']

#Selecionar ufs do Centro Oeste
mensal_aberto_co = mensal_aberto[mensal_aberto['uf'].isin(estados_co)]
mensal_aberto_co
````
