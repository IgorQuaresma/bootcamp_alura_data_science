# Desafio 1: ler esse arquivo direto do wikipedia (html).

```python
#Importar lib de websraping
from bs4 import BeautifulSoup
from urllib.request import urlopen

#objetos da lib
url = 'https://pt.wikipedia.org/wiki/Lista_de_unidades_federativas_do_Brasil_por_popula%C3%A7%C3%A3o#cite_note-IBGE_POP-1'
response = urlopen(url)
html = response.read()
soup = BeautifulSoup(html)


tabela = soup.find('table')

for registro in tabela[1:]:
    
    dicionario_tabela = {}
    contador = 0
    
    
    #Posicao
    dicionario_tabela['posicao'] = contador + 1
    contador += contador

    #UF
    dicionario_tabela['uf'] = registro.find('td',{'align':'left'}).get_text()


    #Populacao
    dicionario_tabela['populacao'] = registro.find('td',{'align':'right'}).get_text()

    #Porcentagem
    dicionario_tabela['porc_pop'] = registro.find('td',{'align':'right'}).findNext('td').get_text()

    #Pais comparavel
    dicionario_tabela['pais_comp'] = registro.find('td',{'align':'right'}).findNext('td').findNext('td').get_text()
    
    
df = pd.DataFrame(dicionario_tabela)
df
    
```
